```python
import os
import random
import shutil
```

```python
def divide_ratio(src, dst, ratio):
    all_files = []
    for root, dirs, files, in os.walk(src):
        for f in files:
            src_path = os.path.join(root, f)
            all_files.append(src_path)
    
    if not all_files:
        print("복사할 파일이 없음")
        return
    
    num_divide = int(len(all_files) * ratio)
    if num_divide == 0:
        print("분할할 파일이 없습니다")
        return
    
    selected = random.sample(all_files, num_divide)
    
    for src_path in selected:
        rel_path = os.path.relpath(src_path, src)
        dst_path = os.path.join(dst, rel_path)
        os.makedirs(os.path.dirname(dst_path), exist_ok=True)
        shutil.copy2(src_path, dst_path)
```

```python
def copy_dir_structure(src, dst):
    for root, dirs, files in os.walk(src):
        rel_path = os.path.relpath(root, src)
        dst_path = os.path.join(dst, rel_path)
        os.makedirs(dst_path, exist_ok = True)
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
SRC = os.getenv("DATA_SRC")
DST = os.getenv("DATA_DST")
copy_dir_structure(src = SRC, dst = DST)
```

```python
divide_ratio(src = SRC, dst = DST, ratio = 0.2)
```
